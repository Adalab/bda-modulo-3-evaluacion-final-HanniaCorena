##      - Numéricas:
       - % ALTO nulos:
         - se imputan con técnicas más avanzadas
       - % BAJO nulos:
         - distribución simétrica: imputación con la media
         - distribución asimétrica: imputación con la mediana

```python
df_clh['Salary'].skew() ## asimetria a la derecha, hay que imputar la mediana
```

```python
import numpy as np
import pandas as pd

# 1️⃣ Forzar conversión TOTAL a numérico
df_clh['Salary'] = pd.to_numeric(df_clh['Salary'], errors='coerce')

# 2️⃣ Convertir negativos a NaN
df_clh.loc[df_clh['Salary'] < 0, 'Salary'] = np.nan

# 3️⃣ Comprobar nulos
print("Nulos antes de imputar:", df_clh['Salary'].isna().sum())

# 4️⃣ Imputar con la mediana
mediana_salary = df_clh['Salary'].median()
df_clh['Salary'] = df_clh['Salary'].fillna(mediana_salary)

# 5️⃣ Verificación FINAL
print("Nulos después de imputar:", df_clh['Salary'].isna().sum())
```

```python
estado_genero = (
    df_clh
    .groupby(['Marital Status', 'Gender'])['Loyalty Number']
    .nunique()
    .reset_index(name='Total Clientes')
)
```

```python
tabla_cruzada = pd.crosstab(
    df_clh['Marital Status'],
    df_clh['Gender']
)
```

```python
tabla_cruzada.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

plt.title('Distribución de clientes por estado civil y género')
plt.xlabel('Estado civil')
plt.ylabel('Número de clientes')
plt.xticks(rotation=45)
plt.legend(title='Género')
plt.show()
```
