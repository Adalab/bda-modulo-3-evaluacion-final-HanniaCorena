## Fase 3: Visualización

Usando las herramientas de visualización que has aprendido durante este módulo, contesta a las siguientes
preguntas usando la mejor gráfica que consideres:

1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año? SE CONTESTA  CON MONTH Y FLIGHTS BOOKED


En todos los ejercicios, no te quedes solo con la representación visual: añade siempre una breve
interpretación en lenguaje natural, como lo haría un analista al presentar resultados a un equipo no
técnico.

```python
vuelos_por_mes = (
    df_cfa
    .groupby('Month')['Flights Booked']
    .sum()
    .reset_index()
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    data=vuelos_por_mes,
    x='Month',
    y='Flights Booked'
)

plt.title('Cantidad de vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Vuelos reservados')
plt.show()
```

```python
sns.scatterplot(
    data=df_cfa,
    x='Distance',
    y='Points Accumulated',
    alpha=0.5
)

plt.title('Relación entre distancia y puntos acumulados')
plt.xlabel('Distancia')
plt.ylabel('Puntos acumulados')
plt.show()
```
