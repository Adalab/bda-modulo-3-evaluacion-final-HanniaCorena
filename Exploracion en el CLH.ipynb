```python
# Calculamos la varianza y la desviación estandard de 'Salary' y las comparamos con la media y mediana.

media_salario = df_clh['Salary'].mean()
mediana_salario = df_clh['Salary'].median()

varianza_salario = df_clh['Salary'].var()
desv_std_salario = df_clh['Salary'].std()

rango = df_clh['Salary'].max() - df_clh['Salary'].min()
cv_salario = desv_std_salario / media_salario * 100
IQR = df_clh['Salary'].quantile(0.75) - df_clh['Salary'].quantile(0.25) # mide la dispersión del 50% central de los datos. Se calcula como la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1).


print(f"La media del salario de los clientes es: {np.round(media_salario, 2)}")
print(f"La mediana del salario de los clientes es: {np.round(mediana_salario, 2)}")
print(f"La varianza  del salario de los clientes es: {np.round(varianza_salario, 2)}")
print(f"La desviación estándar del salario de los clientes es: {np.round(desv_std_salario, 2)}")
print(f"El rango del salario de los clientes es: {np.round(rango, 2)}")
print(f"Coeficiente de Variación: {cv_salario:.2f}%")
print(f"Rango Intercuartílico: {IQR:.2f}")
```
