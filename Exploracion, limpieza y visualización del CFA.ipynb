## Fase 3: Visualización

Usando las herramientas de visualización que has aprendido durante este módulo, contesta a las siguientes
preguntas usando la mejor gráfica que consideres:

1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año? 


Esto lo contestamos con los datos de las columnas Month y Flights Booked



```python
vuelos_por_mes = (
    df_cfa
    .groupby('Month')['Flights Booked']
    .sum()
    .reset_index()
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    data=vuelos_por_mes,
    x='Month',
    y='Flights Booked'
)

plt.title('Cantidad de vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Vuelos reservados')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cfa,
    x='Month',
    y='Flights Booked',
    alpha=0.5
)
plt.title('Distribución de vuelos reservados por mes')
plt.show()
```

Breve interpretación: Más alla de la interpretación sobre la tendencia de mayor vuelos en verano, menores vuelos en invierno y meses mixtos como septiembre y marzo, puedo añadir que los datos parecen lógicos, no hay alguna barra que parezca tener un error.

2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente? 


Para contestarlo, usamos los datos de la columna Distance y Points Accumulated.



```python
sns.scatterplot(
    data=df_cfa,
    x='Distance',
    y='Points Accumulated',
    alpha=0.5
)

plt.title('Relación entre distancia y puntos acumulados')
plt.xlabel('Distancia')
plt.ylabel('Puntos acumulados')
plt.show()
```

```python
correlacion = df_cfa['Distance'].corr(df_cfa['Points Accumulated']) #Da como resultado un float pero está transformado a entero, es un tipo numérico de Numpy
correlacion
```

```python
vars_corr = df_cfa[['Distance', 'Points Accumulated']]

correlation_matrix = vars_corr.corr(method='spearman')

plt.figure(figsize=(6, 4))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Correlación (Spearman) entre Distancia y Puntos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cfa,
    x='Distance',
    y='Points Accumulated',
    alpha=0.5
)
plt.title('Relación entre distancia y puntos acumulados')
plt.show()
```

```python
summary_stats = df_cfa[['Distance', 'Points Accumulated']].agg(['mean', 'median'])
summary_stats
```

Breve descripción: Existe plena relación entre las variables, es decir, entre mas distancia mas puntos se acumulan. Ademas el gráfico no muestra datos dispersos ni valores atípicos.
